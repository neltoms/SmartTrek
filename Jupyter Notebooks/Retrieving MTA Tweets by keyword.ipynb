```python
import twitter
import csv
from datetime import timedelta

#Twitter API credentials
consumer_key = "xxxx"
consumer_secret = "xxxx"
access_key = "xxxx"
access_secret = "xxxx"
```

```python
api = twitter.Api(consumer_key=consumer_key,
                  consumer_secret=consumer_secret,
                  access_token_key=access_key,
                  access_token_secret=access_secret)
```

```python
handle = "NYCTSubway"
keyword = "signal"
from_date = "2018-04-04"
count = "50"
```

```python
results = api.GetSearch(raw_query="q=" + keyword +  
                        "%20from%3A" + handle +  "&src=typd&result_type=recent&since=" + from_date + 
                        "&count=" + count)
```
